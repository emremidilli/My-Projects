# LIBRARIES

```python
import pandas as pd

import numpy as np

import os
```

# CONFIGURATIONS

```python
sOutputSymbol = 'BTCUSD'
sModelType = 'MLP'
```

```python
dfDesign = pd.read_csv('Data\Design_'+ sOutputSymbol +'_'+ sModelType +'.csv', index_col = 1)
```

# IMPORT [Response] VALUES

```python
dfFileNames = pd.DataFrame()
for sRoot, sDirs, aFiles in os.walk("./" + sOutputSymbol):
    for sFileName in aFiles:
        sFilePath = os.path.join(sRoot, sFileName)
        if 'dfPerformance.csv' in sFilePath and sModelType in sRoot :
            dfPerformance = pd.read_csv(sFilePath, index_col=0)
            aFileDetails = sFilePath.split('\\')
            iTrialNr = int(aFileDetails[1].split('_')[1])
            dfDesign.loc[iTrialNr, 'Response'] = dfPerformance.loc['custom_metric', 'value']

dfDesign.drop(dfDesign[dfDesign['Response'].isnull()].index, axis = 0 , inplace = True)
dfDesign
```

```python
dfDesign.to_csv('Data\Design_'+ sOutputSymbol +'_'+ sModelType +'.csv')
```
