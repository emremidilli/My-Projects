# LIBRARIES

```python
import pandas as pd

import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

import os

from scipy.stats import f_oneway
```

# CONFIGURATIONS

```python
np.set_printoptions(suppress=True)
```

```python
pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python
sOutputSymbol = 'BTCUSD'
sModelType = 'MLP'
```

```python
dfCciDesign = pd.read_csv('Data\Cci Design_'+ sOutputSymbol +'_'+ sModelType +'.csv', index_col = 0)
dfCciWithoutBatch = dfCciDesign.drop(['Batch'], axis = 1)
```

# CREATE 1ST ORDER FEATURES

```python
oPolynomialFeatures = PolynomialFeatures(degree=1, interaction_only=True, include_bias=False)
dfAnalysis = pd.DataFrame(data = oPolynomialFeatures.fit_transform(dfCciWithoutBatch))
dfAnalysis.columns = oPolynomialFeatures.get_feature_names_out(dfCciWithoutBatch.columns)
```

# IMPORT [Response] VALUES

```python
dfFileNames = pd.DataFrame()
for sRoot, sDirs, aFiles in os.walk("./BTCUSD"):
    for sFileName in aFiles:
        sFilePath = os.path.join(sRoot, sFileName)
        if 'dfPerformance.csv' in sFilePath:
            dfPerformance = pd.read_csv(sFilePath, index_col=0)
            aFileDetails = sFilePath.split('\\')
            iTrialNr = int(aFileDetails[1].split('_')[1])
            dfAnalysis.loc[iTrialNr, 'Response'] = dfPerformance.loc['custom_metric', 'value']

dfAnalysis.drop(dfAnalysis[dfAnalysis['Response'].isnull()].index, axis = 0 , inplace = True)
dfAnalysis
```

# ANALYSIS OF VARIANCE

```python
aFStatistics, aPValues = f_oneway(dfAnalysis.iloc[0:28:, 0], dfAnalysis.iloc[0:28:, 1], dfAnalysis.iloc[0:28:, 2], dfAnalysis.iloc[0:28:, 3])
```

# FIT TO LINEAR REGRESSION MODEL

```python
oLinearRegression = LinearRegression()
oLinearRegression.fit(
    X = dfAnalysis.iloc[:, :-1],
    y = dfAnalysis.loc[:, 'Response']
)
```
