```python
######################################################## CONSTANTS ########################################################
import CONSTANTS as c

NR_OF_BINS = c.NR_OF_BINS
INTERPLOATION_POINTS = c.INTERPLOATION_POINTS
FFT_AC_COEFFICIENT  = c.FFT_AC_COEFFICIENT
LOWER_POINT = c.LOWER_POINT
UPPER_POINT = c.UPPER_POINT
THRESHOLD = c.THRESHOLD
PREPROCESSED_DATA_PATH = c.PREPROCESSED_DATA_PATH
FEATURE_ENGINEERED_DATA_PATH = c.FEATURE_ENGINEERED_DATA_PATH

######################################################## LIBRARIES ########################################################
import pandas as pd

import numpy as np

from sklearn.preprocessing import MinMaxScaler

import os

i = 0
for oFile in os.walk(PREPROCESSED_DATA_PATH):
    sFolderPath = oFile[0]
    
    if sFolderPath != PREPROCESSED_DATA_PATH:
        sSubFolder = sFolderPath.split('\\')[-1]
        
        sFolderToWrite = r'{}\{}'.format(FEATURE_ENGINEERED_DATA_PATH, sSubFolder)
        
        if os.path.exists(sFolderToWrite) == False:
            os.makedirs(sFolderToWrite)
        
        if all(elem in os.listdir(sFolderToWrite)  for elem in  ['X_ORIGINAL.npy', 'X_ORIGINAL.npy', 'X_ORIGINAL.npy']) == True:
            continue

            
        ######################################################## DATA SOURCE ########################################################
        X_ORIGINAL = np.load(r'{}\X_ORIGINAL.npy'.format(sFolderPath))
        X_TIME_ORIGINAL = np.load(r'{}\X_TIME_ORIGINAL.npy'.format(sFolderPath))
        Y_ORIGINAL = np.load(r'{}\Y_ORIGINAL.npy'.format(sFolderPath))
        
        
        if i ==0:
            X = X_ORIGINAL
            Y = Y_ORIGINAL
            X_TIME = X_TIME_ORIGINAL
        else:
            X = np.concatenate([X, X_ORIGINAL])
            X_TIME = np.concatenate([X_TIME, X_TIME_ORIGINAL])
            Y = np.concatenate([Y, Y_ORIGINAL])            
        
        i = i +1
        
        if i == 3:
            break


X = X[:,:,:,0]
Y = Y[:,:,:,0]
            
######################################################## PREPROCESSING ########################################################


# remove outliers
X[X>UPPER_POINT] = UPPER_POINT
X[X<LOWER_POINT] = LOWER_POINT

Y[Y>UPPER_POINT] = UPPER_POINT
Y[Y<LOWER_POINT] = LOWER_POINT


# XTIME consists of minute, hour, day of week, day of month, month, year
# exclude month and year.
X_TIME = X_TIME[:, :4]

# convert cyclic features to time sinus function
X_TIME = (2 * np.pi * X_TIME)/X_TIME.max(axis = 0)
X_TIME = np.sin(X_TIME)


# scale between 0 and 1
oScaler_X_time = MinMaxScaler()
X_TIME = oScaler_X_time.fit_transform(X_TIME) 

oScaler_X_term = MinMaxScaler()
X = oScaler_X_term.fit_transform(X.reshape(-1, 1)).reshape(X.shape)

oScaler_Y_term = MinMaxScaler()
Y = oScaler_Y_term.fit_transform(Y.reshape(-1, 1)).reshape(Y.shape)
```

```python
######################################################## MODEL ########################################################
import tensorflow as tf


class Block(tf.Module):
    
    def __init__(self, 
                 p_aHiddenArchitecture: list, 
                 p_iFftAcCoeff: int, # we can make this parameter learnable.
                 p_aBinBoundaries: list,
                 p_iNrOfTimeIntervals: int,
                 p_iSizeOfTimeInterval: int, # if 1 time interval represents hour, then size of it is 60.
                 p_iNrOfHeads: int,
                 p_iKeyDims: int,
                 p_iEmbeddingDims: int,
                 p_iBackcastLenght: int,
                 p_iForecastLength: int
                ): 
        super(Block, self).__init__()
        
        
        aLayers = [
            tf.keras.layers.Reshape((1, -1)),
            tf.keras.layers.Dense(p_iEmbeddingDims, name='time_embedding_dense')
        ]
        self.TimeEmbedding = tf.keras.Sequential(aLayers)
        
        
        # input architecture definition for bine distribution by each time interval
        aLayers = [
            tf.keras.layers.Reshape((p_iNrOfTimeIntervals, -1)), 
            tf.keras.layers.TimeDistributed(
                tf.keras.layers.Discretization(
                    bin_boundaries = p_aBinBoundaries,
                    output_mode = 'count',
                    name='time_dist_discritization'
                )
            ),
            tf.keras.layers.Rescaling(scale = 1/p_iSizeOfTimeInterval),
            tf.keras.layers.Dense(p_iEmbeddingDims, name = 'InputBinDistributionByTimeInt_embedding')
        ]
        self.InputBinDistributionByTimeInt = tf.keras.Sequential(aLayers)
        
        # input architecture definition for time interval distribution by each bin
        aLayers = [
            tf.keras.layers.Reshape((p_iNrOfTimeIntervals, -1)), 
            tf.keras.layers.TimeDistributed(
                tf.keras.layers.Discretization(
                    bin_boundaries = p_aBinBoundaries,
                    output_mode = 'count'
                )
            ),
            tf.keras.layers.Rescaling(scale = 1/p_iSizeOfTimeInterval),
            tf.keras.layers.Permute(dims= (2,1)),
            tf.keras.layers.Dense(p_iEmbeddingDims, name = 'InputTimeIntDistributionByBin_embedding')
        ]
        self.InputTimeIntDistributionByBin = tf.keras.Sequential(aLayers)
        
        
        # input architecture definition for fft transformation
        # DESIGN A SEPERATE LAYER WHICH CAN CONVERT INPUTS TO FFT.
        # THIS LAYER SHOULD MERGE REAL AND IMAGINARY PART OF FFT OUTPUT.
        # aLayers = [
        #     tf.keras.layers.Reshape(p_iNrOfTimeIntervals, -1)), 
        #     tf.keras.layers.Lambda(lambda v: tf.gather(tf.signal.fft(tf.cast(v,tf.complex64)),indices= tf.range(self.iFftAcCoeff), axis =2)),
        # ]
        # self.InputFft = tf.keras.Sequential(aLayers)
        
        # concatting all converted inputs
        self.ConcatInputs = tf.keras.layers.Concatenate(axis = 1)
        
        # attention mechanism
        self.AttentionMechanism = tf.keras.layers.MultiHeadAttention(
            num_heads=p_iNrOfHeads, 
            key_dim=p_iKeyDims
        )
        
        
        # mlp stack definition
        aLayers = [
            tf.keras.layers.Flatten()
        ]
        for iHiddenSize in p_aHiddenArchitecture:
            aLayers = aLayers + [
                tf.keras.layers.Dense(iHiddenSize), 
                tf.keras.layers.ReLU()
            ]
        self.MlpStack = tf.keras.Sequential(aLayers)
        
        
        # shared fully connected layer definition
        aLayers = [
            tf.keras.layers.Dense(p_iEmbeddingDims, name = 'SharedFc_dense'),
            tf.keras.layers.ReLU()
        ]
        self.SharedFc = tf.keras.Sequential(aLayers)
        
        
        # backcast layer
        aLayers = [
            tf.keras.layers.Dense(p_iBackcastLenght, name = 'Backcast_dense')
        ]
        self.Backcast = tf.keras.Sequential(aLayers)
        
        
        # forecast layer
        aLayers = [
            tf.keras.layers.Dense(p_iForecastLength, name = 'Forecast_dense')
        ]
        self.Forecast = tf.keras.Sequential(aLayers)        
        
    
    
    # x: should be in the following format -> (nr_of_samples, nr_of_time_steps, nr_of_features)
    # note: time_step and time_interval are different things.
    # time_step = 240 mins
    # time_interval = 4 hours
    def __call__(self, x):
        x_time = x[1]
        x = x[0]
        
        xTimeEmbedded =  self.TimeEmbedding(x_time)
        
        xBinDistByTimeInt = self.InputBinDistributionByTimeInt(x)
        
        xTimeIntDistByBin = self.InputTimeIntDistributionByBin(x)
        
        xMergedFeatures = self.ConcatInputs([xBinDistByTimeInt, xTimeIntDistByBin, xTimeEmbedded])

        xMergedFeatures = self.AttentionMechanism(xMergedFeatures, xMergedFeatures)
        
        xMergedFeatures = self.MlpStack(xMergedFeatures)
            
        xMergedFeatures = self.SharedFc(xMergedFeatures)
        
        xBackward = self.Backcast(xMergedFeatures)
        
        xForward = self.Forecast(xMergedFeatures)
        
        return xBackward, xForward
    
    
class Stack(tf.Module):
    
    def __init__(self, 
                 p_aHiddenArchitecture: list, 
                 p_iFftAcCoeff: int, # we can make this parameter learnable.
                 p_aBinBoundaries: list,
                 p_iNrOfTimeIntervals: int,
                 p_iSizeOfTimeInterval: int, # if 1 time interval represents hour, then size of it is 60.
                 p_iNrOfHeads: int,
                 p_iKeyDims: int,
                 p_iEmbeddingDims: int,
                 p_iBackcastLenght: int,
                 p_iForecastLength: int,
                 p_iNrBlocks: int,
                 p_iIndexOfStack :int
                ): 
        super(Stack, self).__init__()
               
        
        blocks = []
        for i in range(p_iNrBlocks):
            blocks.append(
                Block(
                    p_aHiddenArchitecture =p_aHiddenArchitecture,
                    p_iFftAcCoeff = p_iFftAcCoeff, 
                    p_aBinBoundaries = p_aBinBoundaries,
                    p_iNrOfTimeIntervals= p_iNrOfTimeIntervals,
                    p_iSizeOfTimeInterval=p_iSizeOfTimeInterval,
                    p_iNrOfHeads = p_iNrOfHeads,
                    p_iKeyDims = p_iKeyDims,
                    p_iEmbeddingDims = p_iEmbeddingDims,
                    p_iBackcastLenght=p_iBackcastLenght,
                    p_iForecastLength=p_iForecastLength
                )
            )
            
        self.blocks = blocks
        
        
        
        
    def __call__(self, x):    
        x_time = x[1]
        x = x[0]
        
        
        xStackForecast = []
        
        xStackResidual = x
        
        for oBlock in self.blocks:
            xBackcast, xForecast = oBlock([xStackResidual, x_time])
            
            xStackForecast.append(xForecast)
            
            xStackResidual = tf.keras.layers.subtract([xStackResidual, xBackcast])
            
            
        xStackForecast = tf.keras.layers.Add()(xStackForecast)

        return  xStackResidual, xStackForecast
            
        

        
class ProposedModel(tf.keras.Model):
    
    def __init__(self, 
                 p_aHiddenArchitecture: list, 
                 p_iFftAcCoeff: int, # we can make this parameter learnable.
                 p_aBinBoundaries: list,
                 p_iNrOfTimeIntervals: int,
                 p_iSizeOfTimeInterval: int, # if 1 time interval represents hour, then size of it is 60.
                 p_iNrOfHeads: int,
                 p_iKeyDims: int,
                 p_iEmbeddingDims: int,
                 p_iBackcastLenght: int,
                 p_iForecastLength: int,
                 p_iNrBlocks: int,
                 p_iNrOfStacks :int
                ): 
        super(ProposedModel, self).__init__()
        
        stacks = []
        for i in range(p_iNrOfStacks):
            stacks.append(
                Stack(
                 p_aHiddenArchitecture = p_aHiddenArchitecture, 
                 p_iFftAcCoeff= p_iFftAcCoeff, # we can make this parameter learnable.
                 p_aBinBoundaries = p_aBinBoundaries,
                 p_iNrOfTimeIntervals = p_iNrOfTimeIntervals,
                 p_iSizeOfTimeInterval= p_iSizeOfTimeInterval, # if 1 time interval represents hour, then size of it is 60.
                 p_iNrOfHeads = p_iNrOfHeads,
                 p_iKeyDims = p_iKeyDims,
                 p_iEmbeddingDims = p_iEmbeddingDims,
                 p_iBackcastLenght = p_iBackcastLenght,
                 p_iForecastLength = p_iForecastLength,
                 p_iNrBlocks = p_iNrBlocks,
                 p_iIndexOfStack= i
                )
            )
            
        self.stacks = stacks
        
    
    def call(self, x):
        x_time = x[1]
        x = x[0]
        
        xModelForecast = []
        
        xModelResidual = x
        
        for oStack in self.stacks:
            xBackcast, xForecast = oStack([xModelResidual, x_time])
            
            xModelForecast.append(xForecast)
            
            xModelResidual = tf.keras.layers.subtract([xModelResidual, xBackcast])
            
            
        xModelForecast = tf.keras.layers.Add()(xModelForecast)

        return  xModelForecast        
        
        
        
# compile model      
HIDDEN_ARCHITECTURE = [512,512,512]
BIN_BOUNDARIES = [0.10, 0.20,0.30,0.40,0.50,0.60,0.70,0.80,0.90]
NR_OF_TIME_INTERVAL = 4
SIZE_OF_TIME_INTERVAL = 60
NR_OF_HEADS = 4
KEY_DIMS = 64
EMBEDDING_DIMS = 32
BACKCAST_LENGTH = 240
FORECAST_LENGTH = 60
NR_OF_BLOCKS = 3
NR_OF_STACKS = 2
            
oModel =ProposedModel(
    p_aHiddenArchitecture = HIDDEN_ARCHITECTURE, 
    p_iFftAcCoeff = FFT_AC_COEFFICIENT, # we can make this parameter learnable.
    p_aBinBoundaries = BIN_BOUNDARIES,
    p_iNrOfTimeIntervals = NR_OF_TIME_INTERVAL,
    p_iSizeOfTimeInterval = SIZE_OF_TIME_INTERVAL, # if 1 time interval represents hour, then size of it is 60.
    p_iNrOfHeads = NR_OF_HEADS,
    p_iKeyDims = KEY_DIMS,
    p_iEmbeddingDims=  EMBEDDING_DIMS,
    p_iBackcastLenght = BACKCAST_LENGTH,
    p_iForecastLength = FORECAST_LENGTH,
    p_iNrBlocks = NR_OF_BLOCKS,
    p_iNrOfStacks = NR_OF_STACKS)



# oModel = tf.keras.Model(inputs=[oModel.input], outputs=[oModel.output])



oModel.compile(
    loss = 'mean_squared_error',  
    optimizer = 'adam', 
    metrics = 'mae')
```

```python
iTrainSize=  1000
iTestSize = 50

X_train = X[0:iTrainSize].reshape(-1, 240)
X_TIME_train = X_TIME[0:iTrainSize]
Y_train = Y[0:iTrainSize].reshape(-1, 60)

X_test = X[iTrainSize+1:iTrainSize+1+iTestSize].reshape(-1, 240)
X_TIME_test = X_TIME[iTrainSize+1:iTrainSize+1+iTestSize]
Y_test = Y[iTrainSize+1:iTrainSize+1+iTestSize].reshape(-1, 60)

# Fit model
oModel.fit(
    [X_train, X_TIME_train], 
    Y_train, 
    epochs= 10**2, 
    batch_size=2**8, 
    verbose=1
)
```
